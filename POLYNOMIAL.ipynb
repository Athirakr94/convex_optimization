Import required libraries

```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
Solving for a fixed N and n value. Define the variables required.
```

```python
N=100
n=6
a=cp.Variable((n,1))#a0,a1,....aN-1
M=2*N+1
x=np.linspace(-1,1,M)
xki=[]
for i in range(n):
    xki.append(x**i)
xki=np.array(xki)#nxM
fx=np.power(x,n).reshape(1,M)
```

Define cost function, problem try to solve it for cost value.

```python
cost_fn=cp.sum_squares(fx-a.T@xki)
problem=cp.Problem(cp.Minimize(cost_fn))
problem.solve()
```
